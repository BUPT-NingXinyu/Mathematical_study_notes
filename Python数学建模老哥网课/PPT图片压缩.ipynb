```python
import os
from os import path
from cv2 import cv2

#定义一个函数
def compress_img_in_folder(url):
    files = os.listdir(url)
    for f in files:
        #字符串拼接
        real_url = path.join (url , f)
        imgs = cv2.imread(real_url)
        #图片缩放至原图的1/2
        resize_img = cv2.resize(imgs, (0, 0), fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)
        #重写图片并保存
        cv2.imwrite(real_url, resize_img)
        #打印出来
        print(real_url)

#调用自定义函数
compress_img_in_folder('./img')
```
